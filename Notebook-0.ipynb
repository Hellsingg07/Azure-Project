# **Top 10 Customers by Product (#)**

```sql
Select top 10 (ContactName),COUNT(ProductName) as Product_Count,SUM(o.UnitPrice) as Sum_of_Price from Orders u left join Customers c on u.CustomerID = c.CustomerID left join [Order Details] o on u.OrderID = o.OrderID  left join Products p on o.ProductID = p.ProductID 
where Discount > 0.05
group by ContactName
order by  Product_Count DESC
```

# **Total Sales By Region**

```sql
select RegionDescription, SUM(UnitPrice * Quantity) as Total_Sales from Employees e left JOIN EmployeeTerritories t on e.EmployeeID = t.EmployeeID left join Territories c on t.TerritoryID = c.TerritoryID left join Region r on c.RegionID = r.RegionID left join Orders o on e.EmployeeID = o.EmployeeID left join [Order Details] d on o.OrderID = d.OrderID
group by RegionDescription
order by Total_Sales DESC
```

# **Most Profitable Countries**

```sql
select ShipCountry,Sum(Freight) as Freight,Sum(UnitPrice) as Sales,sum(Quantity) as Total_Units, Sum(UnitPrice * Quantity) as Total_Sales, Sum(UnitPrice * Quantity - Freight) as Profit  from Employees e left JOIN EmployeeTerritories t on e.EmployeeID = t.EmployeeID left join Territories c on t.TerritoryID = c.TerritoryID left join Region r on c.RegionID = r.RegionID left join Orders o on e.EmployeeID = o.EmployeeID left join [Order Details] d on o.OrderID = d.OrderID
group by ShipCountry
Order by Profit DESC
```
